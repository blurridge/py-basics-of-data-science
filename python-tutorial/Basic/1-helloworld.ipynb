# `print()` Function

The `print()` function is very intuitive to use and can display almost every kind of input: *(str, bool, int, float, data structures)*.

There are many ways to format **String Literals** in Python.
*Let's try some!*

```python
######################################
# Different Ways of Printing Strings #
######################################

# Double Quotes
print("Hello World")
# Single Quotes
print('Hello World')

# Multi-line
print('''Hello World''')
print("""Hello World""")

# Concatenation with + Operator with Strings
print('Hello' + ' World')

# Concatenation with , (with implicit space character in between arguments)
print('Hello', 'World')

# f-strings (Format Strings)
print(f"Hello World")
```

# Constants versus Variables

### Constants
> Constant - not changing.
- They are **fixed values** and are usually User-defined in the program and evaluated by the interpreter.
- They usually used as *R-values* **to be assigned** to *Variables*.
- They ***have different types***, which influence the operators that work on them.
- <u>Examples</u> include values like: `21`, `4.20`, `"Hello World"`, `True`, `False`

### Variables
> Variable - able to change or vary.
- They are **variable** and can change according to assignments with the `=` Operator: *L-values*.
- They are **data containers** stored in computer memory.
- associated with an *Identifier* that gives it a user-defined name to access it.
- They are ***Dynamically-typed*** and their `type()` is determined by the value assigned to them.
- They ***can be reassigned to another data type***.

### The `type(arg)` Function
- Returns the type of `arg`.
- Used a lot to ***check what your value/function is evaluated as*** because operators work differently on each type.

```python
print(
    type(5),
    type(5.0),
    type("Hello World"),
    type(True),
    type(False),
    type(None)
)
```

### Reserved Words
- They are words that represent Python's built-in operations/functions and ***CANNOT be used as variable names***.
